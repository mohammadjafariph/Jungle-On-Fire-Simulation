I simulate a jungle on fire model. We have 4 state in system : 
1 - Empty (No tree in that position)
2- Green tree (green)
3- Tree on fire (orange)
4- Burned out tree (black)
Fire starts with one tree in system. then we search for each tree in four main directions (up,down,left,right) for fire . if we found fire in any of these directions we set tree on fire.

```python
import random
from matplotlib import pyplot as plt
font1 = {'family':'serif','color':'blue','size':20}

scale = 40
N = scale * scale
x = []
y = []
status = []
chos = [0,1,1] #---- Chance of being empty is 1/3
for i in range(scale):
    for j in range(scale):
        x.append(i)
        y.append(j)
        choss = random.choice(chos)
        status.append(choss)
```

```python
for i in range(N):
    if status[i] == 1:
        plt.scatter(x[i],y[i],color='green')
plt.savefig('outjungle/start.png')
```

```python
ir = random.randint(0,100)
while status[ir] == 0:
    ir = random.randint(0,100)
    

    
    
status[ir] = 2 #--Fire started
xy = []
for i in range(N):
    xy.append((x[i],y[i]))
burned = []
for i in range(50):
    
    brncount = 0
    for j in range(N):
        
        if status[j] == 2:
            status[j] = 3
            
        elif status[j] == 1 :
            if x[j] > 0 and x[j]< (scale-1) and y[j] > 0 and y[j] < (scale-1):
                neighbors = [(x[j],y[j]+1),(x[j],y[j]-1),(x[j]+1,y[j]),(x[j]-1,y[j])] # cheking fire on neigbor trees in up,down,left and right
            elif x[j] == 0 and y[j] == 0:
                neighbors = [(x[j],y[j]+1),(x[j]+1,y[j])]
            elif x[j] == 0 and y[j] == (scale-1) :
                 neighbors = [(x[j],y[j]-1),(x[j]+1,y[j])]
            elif x[j] == (scale-1) and y[j] == 0 :
                 neighbors = [(x[j],y[j]+1),(x[j]-1,y[j])]
            elif x[j] == (scale-1) and y[j] == (scale-1):
                 neighbors = [(x[j],y[j]-1),(x[j]-1,y[j])]
            
            elif x[j] == 0 :
                 neighbors = [(x[j],y[j]+1),(x[j],y[j]-1),(x[j]+1,y[j])]
            elif x[j] == (scale-1) :
                 neighbors = [(x[j],y[j]+1),(x[j],y[j]-1),(x[j]-1,y[j])]
            elif y[j] == 0 :
                neighbors = [(x[j],y[j]+1),(x[j]+1,y[j]),(x[j]-1,y[j])]
            elif y[j] == (scale-1) :
                neighbors = [(x[j],y[j]-1),(x[j]+1,y[j]),(x[j]-1,y[j])]                    
            for neighbor in neighbors:
                if status[xy.index(neighbor)] == 2 :
                    status[j] = 2
                    
                    
    
    for k in range(N):
        if status[k] == 1:
            plt.scatter(x[k],y[k],color='green')
        elif status[k] == 2:
            plt.scatter(x[k],y[k],color='orange')
            brncount = brncount + 1
        elif status[k] == 3:
            plt.scatter(x[k],y[k],color='black')
            
                
    burned.append(brncount)
    plt.title("Jungle on fire - By Mohammad Jafari")
    plt.xlabel("t = {}".format(i) , fontdict = font1)
    plt.savefig('outjungle/{}.png'.format(i))

    
            
            
        
```

```python
survived = 0
for i in status:
    if i == 1 :
        
        survived = survived + 1
    
for k in range(N):
    if status[k] == 1:
        plt.scatter(x[k],y[k],color='green')
    elif status[k] == 2:
        plt.scatter(x[k],y[k],color='orange')
        
    elif status[k] == 3:
        plt.scatter(x[k],y[k],color='black')


font2 = {'family':'serif','color':'red','size':15}
plt.title("Jungle on fire - By Mohammad Jafari")
plt.xlabel("t = 49", fontdict = font1)
plt.ylabel("Survived Trees = {}".format(survived), fontdict = font2)
plt.savefig('outjungle/end.png')

    
```

```python
plt.title("Burned trees in each step")
plt.xlabel("t")
plt.ylabel("Burned trees")
plt.plot(range(50),burned)
plt.savefig('outjungle/stat.png')
```
